```python
wiki = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
page = urllib.request.urlopen(wiki)
```

```python
from bs4 import BeautifulSoup
```

```python
formated = BeautifulSoup(page)
print (formated)
```

```python
all_tables = formated.find_all('table')
print (all_tables)
```

```python
my_table=formated.find('table', class_='wikitable sortable')
print (my_table)
```

```python
A=[]
B=[]
C=[]
for row in my_table.findAll("tr"):
    cells = row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
print (A)
print (B)
print (C)
```

```python
for n, i in enumerate(B):
    if i == 'Not assigned':
        B[n] = 'NaN'
```

```python
for n, i in enumerate(C):
    if i == 'Not assigned\n':
        C[n] = 'NaN'
```

```python
for n, i in enumerate(A):
    if i == 'Not assigned':
        A[n] = 'NaN'
```

```python
D = []
for i in C:
    D.append(i.strip())
```

```python
l = {'Postcode': A, 'Borough': B, 'Neighborhood': C}
df = pd.DataFrame(data=l)
df
```

```python
df = df[df.Borough != 'NaN']
df.head(7)
```

```python
df.reset_index(drop=True, inplace=True)
df.head(7)
```

```python
df['Neighborhood'].replace('NaN', df['Borough'], inplace=True)
df.head(7)
```
